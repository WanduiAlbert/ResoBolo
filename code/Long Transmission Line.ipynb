```python
import matplotlib.pyplot as plt
% matplotlib inline

from astropy.constants import c
import astropy.units as u
```

```python
Z0 = 50 * u.Ohm
Zc = 29 * u.Ohm
e_eff = 3.35
v_ph = c/e_eff**0.5
```

```python
f  = np.r_[200:1000:1000j] * u.MHz
beta = 2*pi*f/v_ph
l = 2016200 * u.um
x = (beta * l).to(1)
```

```python
S21 = (2 * Z0 * Zc)/((2 * Z0 * Zc)*np.cos(x * u.rad) + 1j * ( Z0**2 +  Zc**2) * np.sin(x * u.rad))
S11 = (1j *(Zc**2 - Z0**2))/((2 * Z0 * Zc)*np.cos(x * u.rad) + 1j * ( Z0**2 +  Zc**2) * np.sin(x * u.rad))
```

```python
S11_mag = 20 * np.log10(np.abs(S11))
S21_mag = 20 * np.log10(np.abs(S21))
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(f, np.real(S21), 'r-', label='Real')
ax.plot(f, np.imag(S21), 'b-', label='Imag')
ax.set_xlabel(r'Frequency [MHz]')
ax.set_ylabel(r'S21')
ax.legend(loc='best')
ax.axis('tight')
ax.grid(which='both');
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(f, S21_mag, 'b-')
ax.set_xlabel(r'Frequency [MHz]')
ax.set_ylabel(r'S21')
ax.axis('tight')
ax.grid(which='both');
```
