```python
import numpy as np
from scipy import stats

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
np.random.seed(25)
```

```python
alpha, beta = [1, 2.5]
x = np.random.randn(10)
x = np.sort(x)
n = np.random.randn(10) * 0.2
y = alpha + beta *x + n
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(x, y);
```

```python
A = np.vander(x, 2, increasing=True)
C = np.diag(np.ones_like(x)*np.std(n))
Cinv = np.linalg.inv(C)
F = np.dot(A.T, np.dot(Cinv, A))
```

```python
thetap = np.dot(np.linalg.inv(F), np.dot(A.T, np.dot(Cinv, y)))
thetap
```

```python
H = np.dot(A, np.dot(np.linalg.inv(F), np.dot(A.T, Cinv)))
yp = np.dot(H, y)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, y);
ax.plot(x, yp, 'r.')
ax.plot(x, np.dot(A, thetap),'k')
ax.grid(which='both');
```
