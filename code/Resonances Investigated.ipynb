```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Qc = 4e5
Qr = Qc/2
x = np.r_[-10:10:1000j]
S21 = 1 - (Qr/Qc)/(1 + 1j * 2 * Qr * x)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, np.abs(S21))
ax.set_xlabel('x')
ax.set_ylabel(r'$S_{21}$')
ax.axis('tight')
ax.grid()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.semilogx(x + 1, np.abs(S21))
ax.set_xlabel('x')
ax.set_ylabel(r'$S_{21}$')
ax.axis('tight')
ax.grid()
```
