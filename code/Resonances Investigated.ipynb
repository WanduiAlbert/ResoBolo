```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, np.abs(S21), 'b')
# ax.plot(x, phi_prime)
ax2 = ax.twinx()
ax2.plot(x, phi_prime, 'r')
ax.set_xlabel('x')
ax.set_ylabel(r'$S_{21}$')
ax2.set_ylabel(r'$\frac{\mathrm d \phi}{\mathrm d x}$')
# ax.axis('tight')
ax.set_xlim(left=-0.1, right=0.1)
# ax2.axis('tight')
ax.grid()
```
