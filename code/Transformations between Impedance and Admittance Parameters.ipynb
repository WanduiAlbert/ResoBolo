```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
%matplotlib inline
```

```python
nH = 1e-9
pF = 1e-12
MHz = 1e6
L = 10*nH
C = 5*pF
fr = 1/2/pi/(L*C)**0.5/MHz
```

```python
Q = 100
R = (2*pi*fr*MHz*L/Q)
Z0 = 50
```

```python
x = np.r_[-3:3:1000j]
Z = R*(1+2j*Q*x)
S21 = (2*Z)/(2*Z + Z0)
```

```python
print (R)
print (fr)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x, np.abs(Z))
ax.axis('tight')
ax.grid(which='both');
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x, np.abs(S21))
ax.axis('tight')
ax.grid(which='both');
```
