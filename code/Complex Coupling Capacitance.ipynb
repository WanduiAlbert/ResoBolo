```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def S21(x, theta):
    Qi, Qe = theta
    Qc = np.real(Qe)
    Qr = 1/(1/Qi + 1/Qc)
    return 1 - (Qr/Qe)/(1 + 1j * 2 * Qr * x)
```

```python
x = np.r_[-5:5:1000j]*1e-4
Qc = 40000
Qi = 80000
Qe_im = [-10000, -1000, -100,-10, -5, -1, -0.1, 0, 0.1, 1, 5, 10, 100, 1000, 10000]
transmission = S21(x, [Qi, Qc])
```

```python
fig, ax = plt.subplots(figsize=(10,10))
for Qim in Qe_im:
    Qe = Qc + 1j * Qim
    transmission = S21(x, [Qi, Qe])
    mag = 10*np.log10(np.abs(transmission))
    ax.plot(x*1e4, mag, label="%3.1f"%Qim)
ax.set_xlabel('Fractional Frequency x')
ax.set_ylabel(r'$|S_{21}|$')
lg = ax.legend(loc='lower right', bbox_to_anchor=(1.35, 0.1))
lg.set_title(r'$\operatorname{\mathbb{I}m}(Q_{e})$', prop={'size':20})
ax.grid(which='both')
ax.axis('tight');
```

```python
fig, ax = plt.subplots(figsize=(10,10))
for Qim in Qe_im:
    Qe = Qc + 1j * Qim
    transmission = S21(x, [Qi, Qe])
    phase = np.unwrap(np.angle(transmission))
    ax.plot(x*1e5, phase, label="%3.1f"%Qim)
ax.set_xlabel('Fractional Frequency x')
ax.set_ylabel('Phase')
lg = ax.legend(loc='lower right', bbox_to_anchor=(1.35, 0.1))
lg.set_title(r'$\operatorname{\mathbb{I}m}(Q_{e})$', prop={'size':20})
ax.grid(which='both')
ax.axis('tight');
```
