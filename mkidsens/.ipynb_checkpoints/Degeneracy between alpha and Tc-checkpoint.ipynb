```python
import numpy as np
from scipy.constants import h,k,c,e
import scipy.special as special
import matplotlib.pyplot as plt
from math import pi
%matplotlib inline
```

```python
K0 = lambda x: special.kn(0, x)
I0 = lambda x: special.iv(0, x)

um = 1e-6
nm = 1e-9
ppm = 1e6
mK = 1e-3

Fd0 = 5.97e-5
Tc0 = 1.284
Delta0 = 1.763*k*Tc0
alphak0 = 0.375
N0 = 1.72e10/um**3/e
R = 10.7*um**3#/s
nqp_star = 518/um**3
Vsc = 16200*um**3*50*nm
```

```python
def model_xMB(T, f0, Tc, alphak, nfloor):
	Delta = 1.763*k*Tc
    nfloor /= um**3
    nth = 2*N0*np.sqrt(2*Delta*pi*k*T)*np.exp(-Delta/k/T)
    nqp = np.sqrt((nth + nqp_star)**2 + nfloor**2) - nqp_star
    lwa = (pi*(4*N0*Delta/nqp)**2)
    Teffs = ((2*Delta/kB)/special.lambertw(lwa)).real
	eta = h*f0*1e6/(2*k*Teffs)
	S2 = 1 + np.sqrt(2*Delta/(pi*k*Teffs))*np.exp(-eta)*I0(eta)
	xmb = -alphak*S2*nqp/(4*N0*Delta)
	return xmb

def model_dQMB(T, f0, Tc, alphak, nfloor):
	Delta = 1.763*k*Tc
    nfloor /= um**3
    nth = 2*N0*np.sqrt(2*Delta*pi*k*T)*np.exp(-Delta/k/T)
    nqp = np.sqrt((nth + nqp_star)**2 + nfloor**2) - nqp_star
    lwa = (pi*(4*N0*Delta/nqp)**2)
    Teffs = ((2*Delta/kB)/special.lambertw(lwa)).real
	eta = h*f0*1e6/(2*k*Teffs)
	S1 = (2/pi)*np.sqrt(2*Delta/(pi*k*Teffs))*np.sinh(eta)*K0(eta)
	dQmb = alphak*S1*nqp/(2*N0*Delta)
	return dQmb

def get_Teff(T, f0, Tc, alphak, nfloor):
    Delta = 1.763*k*Tc
    nfloor /= um**3
    nth = 2*N0*np.sqrt(2*Delta*pi*k*T)*np.exp(-Delta/k/T)
    nqp = np.sqrt((nth + nqp_star)**2 + nfloor**2) - nqp_star
    lwa = (pi*(4*N0*Delta/nqp)**2)
    Teffs = ((2*Delta/kB)/special.lambertw(lwa)).real
    return Teffs
```

```python
T = np.r_[80:300:1000j]*1e-3
Tcs = np.r_[1.1:1.5:10j]
alphaks = np.r_[0:1:10j]
X,Y,Z = np.meshgrid(T, Tcs, alphaks)
```

```python
f0 = 300 #MHz
x_fit = model_xMB(X, f0, Y, Z)*ppm
dQ_fit = model_dQMB(X, f0, Y, Z)*ppm
T_eff = get_Teff(X, f0, Y, Z)
```

```python
f0 = 300 #MHz
fig, ax = plt.subplots(figsize=(10,10))
for alph in alphaks:
    T_eff = get_Teff(T, f0, Tc0, alph)
    ax.plot(T/mK, T_eff/mK, label="alphak = %1.2f"%(alph))
ax.grid()
ax.legend(loc='lower left')
ax.set_xlabel('Temperature [mK]')
ax.set_ylabel('T eff [mK]');
```
